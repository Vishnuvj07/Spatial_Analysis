```python
import matplotlib
import geopandas as gpd
import contextily as ctx
```

```python
deaths_df = gpd.read_file('dataset/SnowGIS_v2/SnowGIS/Cholera_Deaths.shp')
pumps_df = gpd.read_file('dataset/SnowGIS_v2/SnowGIS/Pumps.shp')

deaths_df.head()
```

```python
ax = deaths_df.plot(column='Count', figsize=(15, 15), alpha=0.5, edgecolor='k', legend=True)
pumps_df.plot(ax=ax, marker='x', color='red', markersize=50)
```

```python
ax = deaths_df.plot(column='Count', figsize=(15, 15), alpha=0.5, edgecolor='k', legend=True)
pumps_df.plot(ax=ax, marker='x', color='red', markersize=50)

ctx.add_basemap(
    ax,
    # CRS definition. Without the line below, the map stops making sense
    crs=deaths_df.crs.to_string(),
)
```

```python
ax = deaths_df.plot(column='Count', figsize=(15, 15), alpha=0.5, edgecolor='k', legend=True)
pumps_df.plot(ax=ax, marker='x', color='red', markersize=50);

ctx.add_basemap(ax,
    crs=deaths_df.crs.to_string(),
    # Using the original map, hand-drawn by Snow
    source="dataset/SnowGIS_v2/SnowGIS/SnowMap.tif"
)
```
